The programs we have done so far followed what is called a _squential structure_, that is, the execution of the program was done instruction by instruction from the first to the last of the program, one after the other.

This is not the only possibility in programming and in fact programs usually result from the combination of three types of structures:

- Sequential structure, where a series of instructions are executed one after the other, in the same order in which they are written.
- Alternative structure, where one set of instructions is executed or another depending on whether a condition is met or not.
- Repetitive structure, where the same set of instructions is executed several times.

# 3.2 Alternative structure

The idea is very simple, it is necessary to indicate to the computer that in case a condition is fulfilled (`True`) the instructions to execute will be some and in opposite (` False`) others.

The alternative structure in Python is coded as follows:

```Python
if condition:
    Action 1 # instruction or instructions to be executed IF THE CONDITION IS FULFILLED
else:
    Action 2 # instruction or instructions to execute Otherwise
```

Programming languages are characterized by a very strict syntax and therefore it is important that we pay attention to how the structure has been written:

- The `if` statement indicates that this is an _alternative structure_
- The condition is specified below
- The colon `:` **is required**
- The instruction or sequence of instructions to execute **in case the condition is met** (`True`). Note that these are separated from the left margin. Specifically, **4 spaces**. We call this separation **indentation** and it is also a mandatory syntax rule.
- Then the keyword `else`, which means _otherwise_.
- Again, the colon `:` **is required**
- The instruction or series of instructions to be executed **in case the condition is NOT fulfilled** (`False`). Note that these instructions should also be indented.

## 3.2.2. Variants of the alternative structure

The alternative structure allows different variants. Thus, for example, there may be different possible options or we may only be interested in a series of sentences being executed if a condition is met and nothing else.

### 3.2.2.1. Multiple conditions

In the event that it is necessary to fill a range of possibilities and that it is necessary to establish multiple conditions, the most natural solution that can come to mind is to put alternative structures inside each other. For example:

```python
if condition1:
    Action_1
else:
    if condition2:
        Action_2
    else:
        if condition3:
            Action_3
        else:
            Action_n
Action_x
```
where actions mean one or more instructions.

Let's try to understand how it works in this example. In case the `condition1` is met, the first block is executed (Action_1); otherwise, if the `condition2` is met, the second block will be executed (Action_2); otherwise if the `condition3` is met then the instructions in the third block will be executed (Action_3). And in case any of the above conditions are true then the computer will execute the Action_n instructions.

When one of the previous blocks has been executed, it will move to `Action_x`, which is no longer part of the alternate structure.

The Python language also allows the use of the `elif` command, which would become a _concatenation of` else` and `if`_. An equivalent structure to the above would be the following, which is easier to read:

```python
if condition1:
     Action_1
elif condition2:
     Action_2
elif condition3:
     Action_3
else:
     Action_n
Action_x
```

In this case, as before, the instructions corresponding to the first condition that is fulfilled will be executed, in order from top to bottom, and in case none of them are fulfilled, those corresponding to the `else`.

Once one of these blocks of instructions has been executed, it will be moved to Action_x, which is no longer part of the alternate structure.

In both cases, it must be borne in mind that the **conditions must be well-defined from a logical point of view** since the interpreter will not check for possible compatibility. That is, if two conditions are met simultaneously (e.g. the first and the second) the first one that is verified will be executed and not the next one!

### 3.2.2.2. Otherwise ... do nothing

Other times we will find ourselves in a situation where we only want a series of instructions to be executed if the condition is verified and in case it is not met, the program does nothing. In this case, the `else:` statement and its corresponding block can be ignored:
```python
if condition:
     Action
Action_x
```
In the event that the condition is met, the instructions corresponding to the Action will be executed. If the condition is not met, the computer ignores the block and goes directly to the first Action_x statement.

We can see it in this program fragment that changes the sign of the value of the variable x only in the case in which it has a negative value.
```Python
x = float (input ("Enter a real value:"))
if x <0:
     x = -x
print (x)
```
